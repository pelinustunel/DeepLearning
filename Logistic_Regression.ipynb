```python
low_var_cols = []
for col in dataset.columns:
    if dataset[col].var() == 0:
        low_var_cols.append(col)
print(low_var_cols)
```

```python
cols_to_drop = ['Bankrupt?'] + low_var_cols
x = dataset.drop(cols_to_drop, axis=1)
y = dataset['Bankrupt?']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

# Max-iter kısmını yazmazsak default olarak 100 kabul ediyo
log_reg = LogisticRegression(random_state = 42, max_iter = 300)
log_reg.fit(x_train_scaled, y_train)
```

```python
y_pred_ir = log_reg.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

cs = confusion_matrix(y_test, y_pred_ir)
print(cs)
```

```python
print(classification_report(y_test, y_pred_ir))
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred_ir))
```
